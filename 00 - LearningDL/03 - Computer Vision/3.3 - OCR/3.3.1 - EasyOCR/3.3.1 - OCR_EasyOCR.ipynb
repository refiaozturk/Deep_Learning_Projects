```python
import cv2
import matplotlib.pyplot as plt
```

```python
import easyocr
```

```python
reader = easyocr.Reader(['tr', 'en'], gpu=False)  
```

**reader = easyocr.Reader(['tr', 'en'], gpu=False)**: We create a reader using the easyocr library. This reader is set to recognize both Turkish and English languages (['tr', 'en']).  

The **gpu=False** parameter specifies that the GPU will not be used, so a CPU-based solution is utilized instead.

# <font color='mediumseagreen'> <b>Image 1</b><font color='black'>  

```python
img = cv2.imread('gazi.jpg')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb);
```

```python
results = reader.readtext(img, detail=1, paragraph=True) 
```

- **results_paragraph = reader.readtext(img, detail=1, paragraph=True):** Reads the text from the image using the created reader.
- **detail=1** ensures that detailed results are obtained, and
- **paragraph=True** retrieves the text as a paragraph.

```python
results
```

```python
results = reader.readtext(img, detail=1, paragraph=False) 
```

```python
results
```

```python
results = reader.readtext(img)
```
