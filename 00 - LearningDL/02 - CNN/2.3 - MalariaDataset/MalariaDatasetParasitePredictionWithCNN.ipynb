**Malaria Cell Images Dataset**

The **Malaria Cell Images Dataset** is used in medical image processing and artificial intelligence research. This dataset can be utilized for training disease classification models. It contains cell images that show changes in red blood cells caused by malaria, a disease triggered by protozoan parasites known as *Plasmodium*.

The dataset is provided by the **National Institutes of Health (NIH)** and contains **27,558** cell images. Among these, **13,779** images contain malaria parasites, while the others do not. The dataset consists of two classes: **Malaria-Infected** and **Uninfected**.

**Applications**

This dataset can be used for:
- **Classification tasks**  
- **Object detection**  
- Other **machine learning** tasks  

It can also be applied in other medical imaging domains like **Computed Tomography (CT)** and **Magnetic Resonance Imaging (MRI)**. The dataset offers a wide range of applications where AI algorithms can contribute to medical diagnosis and treatment processes.

**Data Collection**

The malaria cell images were captured using a **microscope**. These images are thin smears of red blood cells, showing the effects of malaria caused by *Plasmodium* parasites. Microscopic images are critical for medical research and diagnostics, aiding in disease detection and treatment.

Additionally, the dataset supports the development of AI algorithms for medical diagnosis and treatment.

**Digitalization Process**

The **Malaria Cell Images Dataset** was created through the **digitization** of these microscopic images. Digitization refers to converting images into numerical data formats, making them suitable for machine learning algorithms.

In summary, the malaria cell images are thin smears of human blood cells photographed under a microscope, highlighting the effects of malaria. These images are valuable in medical research and the diagnosis and treatment of the disease.

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
````

````python
image_gen = ImageDataGenerator(rotation_range=15, 
                               width_shift_range=0.10, 
                               height_shift_range=0.10, 
                               #rescale=1/255,           ==> data is pre-scaled
                               shear_range=0.1, 
                               zoom_range=0.1, 
                               horizontal_flip=True, 
                               fill_mode='nearest')
````

This process does not duplicate the existing data in a concrete way; instead, it provides diversity through various transformations before feeding the data into the model.

**rotation_range:** The range of random rotation angles (in degrees). For example, when `rotation_range=15` is set, images can be randomly rotated between -15 and +15 degrees.

**width_shift_range:** The proportion of horizontal shift for the image. This allows the image to be randomly shifted along the horizontal axis. The value should be a ratio between 0 and 1. For example, with `width_shift_range=0.10`, images can be shifted horizontally by up to 10% of their width.

**height_shift_range:** The proportion of vertical shift for the image. This allows the image to be randomly shifted along the vertical axis. The value should be a ratio between 0 and 1. For example, with `height_shift_range=0.10`, images can be shifted vertically by up to 10% of their height.

**shear_range:** The degree of shearing effect (in radians). This allows the image to be randomly distorted along the horizontal or vertical axes. For example, with `shear_range=0.1`, images can be randomly sheared by up to 0.1 radians.

**zoom_range:** The zoom range for the image. This allows the image to be randomly zoomed in or out. For example, with `zoom_range=0.1`, images can be randomly zoomed between 90% and 110% of their original size.

**horizontal_flip:** Enables or disables random horizontal flipping. If set to `True`, images can be randomly flipped horizontally.

**fill_mode:** Determines how to fill in the empty pixels that appear after transformations like rotation or shifting. The `'nearest'` option fills these gaps with the nearest pixel values.

# <font color='pink'> <b>Modeling</b><font color='black'>  

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
````

````python
from tensorflow.keras.layers import BatchNormalization
````

````python
image_shape = (130,130,3)
````

````python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3,3),input_shape=image_shape, padding='same', activation='relu',))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu',))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

#model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
#model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
#model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
#model.add(MaxPooling2D(pool_size=(2, 2)))

#model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
#model.add(MaxPooling2D(pool_size=(2, 2)))

#model.add(Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu',))
#model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Flatten())


model.add(Dense(64))
model.add(Activation('relu'))

model.add(Dropout(0.3))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
````

````python
model.summary()
````

## <font color='palevioletred'> <b>Training the Model</b><font color='black'>  

````python
#help(image_gen.flow_from_directory)
````

````python
batch_size = 32
train_image_gen = image_gen.flow_from_directory(train_path,
                                               target_size=(130,130),
                                               color_mode='rgb',
                                               batch_size=batch_size,
                                               class_mode='binary', shuffle=True, seed=42)
````

The `image_gen.flow_from_directory()` method creates a data flow for training data by utilizing a folder structure where the data is classified into subdirectories within a main directory on disk.

**Parameters of this method:**

- **directory:** The path to the directory containing the training data.
- **target_size:** The target dimensions to which the images will be resized.
- **color_mode:** The mode of color channels in the images, e.g., `"rgb"` or `"grayscale"`.
- **batch_size:** The size of the data batch to be generated at each iteration.
- **class_mode:** The classification mode, e.g., `"binary"` or `"categorical"`.
- **shuffle:** Determines whether the data should be shuffled or not.
- **seed:** A predefined random seed value used for data shuffling to ensure reproducibility.

This method generates a new batch of data with each call, allowing the model to work with fresh data at every training step. Additionally, any data augmentation operations defined with `ImageDataGenerator` are dynamically applied during each call. This enables the model to access more varied data during training, improving its ability to generalize.

````python
image_gen_test = ImageDataGenerator()

test_image_gen = image_gen_test.flow_from_directory(test_path,

                                                   target_size=(130,130),
                                               
                                                   color_mode='rgb',
                                               
                                                   batch_size=batch_size,
                                               
                                                   class_mode='binary', shuffle=False, seed=42)
````
